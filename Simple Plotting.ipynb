```python
#Import the necessary packages
import matplotlib.pyplot as plt

#Plotting a linear function

#Create a linear array
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

#Plot the linear function
#In order to change the the color I used 'r' to change the color to red
#The marker '.' adds small dots to the function
#The linestyle '--' changes from a conncted line to a broken dashed line
#Alpha just decreases the opacity of the line
plt.plot(x, y, 'r', marker = '.', linestyle = '--', alpha = .8, label = "x")
plt.plot([2, 4, 6, 8, 10], label = "2x")
plt.title('Linear Function')

plt.xlabel('x-axis')
plt.ylabel('y-axis')

#I added a legend to differentiate the difference between the two functions
#loc just positions the legend map where 0 is the best placement for the map
plt.legend(loc = 0, fontsize = 'large', title = "Legend")
plt.show()
```

```python
#Plotting other functions

#Instead of manually making an array of numbers for x I can use numpy to generate a list for me!
import numpy as np

x = np.array(range(10))

y1 = x ** 2

y2 = x ** 1/2

y3 = x ** 3

plt.plot(x, y1, 'c', label = '$2^x$', marker = '*')
plt.plot(x, y2, 'm', label = r'$x^\frac{1}{2}$', marker = '>')
plt.plot(x, y3, 'g', label = '$2^3$', marker = '.')
plt.legend(fontsize = 'large')
plt.show()
```

```python
#Other types of plots

#Making a bar chart

x = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

y = [5, 7, 5, 8, 4, 10, 9]

plt.bar(x, y)
plt.xlabel('Days of the week')
plt.ylabel('Hours of sleep')
plt.title('Hours of sleep per day')
```

```python
#Making a Scatter plot

randData = np.random.rand(50, 2)

randData
x = randData[:,0]
y = randData[:,1]

plt.scatter(x, y, marker = '*')
```

```python
#Making a Pie chart

games = 'Black Desert Online', 'Minecraft', 'Home Sweet Home', 'Final Fantasy XII'
percentPlayed = [30, 40, 5, 25]
explode = (0, .1, 0, 0)

plt.pie(percentPlayed, explode = explode, labels = games, autopct = '%1.0f%%')
```
